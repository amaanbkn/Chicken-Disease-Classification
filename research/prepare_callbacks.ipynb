```python
import os
import time
import tensorflow as tf
from cnnClassifier.entity.config_entity import PrepareCallbacksConfig

class PrepareCallback:
    def __init__(self, config: PrepareCallbacksConfig):
        self.config = config

    def _create_tb_callbacks(self):
        timestamp = time.strftime("%Y-%m-%d-%H-%M-%S")
        tb_root = str(self.config.tensorboard_root_log_dir)
        tb_run_dir = os.path.join(tb_root, f"tb_logs_at_{timestamp}")
        return tf.keras.callbacks.TensorBoard(log_dir=tb_run_dir)

    def _create_ckpt_callbacks(self):
        ckpt_path = str(self.config.checkpoint_model_filepath)
        # If a directory was provided, append a file name
        if ckpt_path.endswith(("/", "\\")) or os.path.isdir(ckpt_path):
            ckpt_path = os.path.join(ckpt_path, "model.ckpt.keras")
        # Ensure extension so Keras doesn't guess and call endswith on non-str later
        if not (ckpt_path.endswith(".keras") or ckpt_path.endswith(".h5")):
            ckpt_path += ".keras"
        return tf.keras.callbacks.ModelCheckpoint(
            filepath=ckpt_path,
            save_best_only=True,
        )

    def get_tb_ckpt_callbacks(self):
        return [self._create_tb_callbacks(), self._create_ckpt_callbacks()]
```
